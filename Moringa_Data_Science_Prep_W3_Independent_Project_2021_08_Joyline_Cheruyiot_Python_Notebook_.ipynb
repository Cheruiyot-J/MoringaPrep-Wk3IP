#### Merging telecom dataset with Cell Geo Dataset

```python
# Merging using siteIDs and SiteCodes
final_df = pd.merge( tel_123, cell_geo, left_on='SITE_ID' , right_on='SITE_CODE',validate='m:1', how='inner')
final_df.head()
```

```python
df_datetime = final_df['DATE_TIME'].apply(lambda x: pd.Series([i for i in reversed(x.lower().split(' '))]))
df_datetime.columns = ['TIME', 'DATE']
df_datetime
```

```python
df_time = df_datetime['TIME'].apply(lambda x: pd.Series([i for i in reversed(x.lower().split(':'))]))
df_time.columns = ['s','m','h']
df_time['h'] = pd.to_numeric(df_time['h'])
df_time['m'].unique()
```

```python
final_df = df_datetime.merge(final_df, left_index=True, right_index=True)
final_df

#merging splitted time to the final dataset
final_df = df_time.merge(final_df, left_index=True, right_index=True)
final_df
```

```python
final_df = final_df.drop('DATE_TIME', axis=1)
final_df
```
